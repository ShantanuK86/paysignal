```python
import pandas as pd

df = pd.read_csv('../data/ds_salaries.csv')

print("Shape:", df.shape)           # rows, columns
print("\nColumns:\n", df.columns.tolist())
print("\nFirst 3 rows:")
df.head(3)
```

```python
import matplotlib
matplotlib.use('Agg')  # safe for notebook
import matplotlib.pyplot as plt

print("Salary stats:")
print(df['salary_in_usd'].describe())

print("\nMin salary:  $", df['salary_in_usd'].min())
print("Max salary:  $", df['salary_in_usd'].max())
print("Mean salary: $", round(df['salary_in_usd'].mean()))
```

```python
print("Missing values per column:")
print(df.isnull().sum())

print("\nData types:")
print(df.dtypes)
```

```python
print("Experience levels:")
print(df['experience_level'].value_counts())

print("\nAverage salary by experience:")
print(df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False).round(0))
```

```python
print("Remote ratio distribution:")
print(df['remote_ratio'].value_counts())

print("\nAverage salary by remote ratio:")
print(df.groupby('remote_ratio')['salary_in_usd'].mean().round(0))

print("\nSalary trends by year:")
print(df.groupby('work_year')['salary_in_usd'].mean().round(0))
```

```python
print("Company size vs salary:")
print(df.groupby('company_size')['salary_in_usd'].mean().round(0))

# S = small, M = medium, L = large
```

```python
top_titles = (df.groupby('job_title')['salary_in_usd']
               .mean()
               .sort_values(ascending=False)
               .head(15)
               .round(0))

print("Top 15 paying roles:")
print(top_titles)
```

```python
# This is your "data contract" — columns we'll use for training
selected_features = [
    'work_year',
    'experience_level',   # EN, MI, SE, EX
    'job_title',          # 50+ roles
    'remote_ratio',       # 0, 50, 100
    'company_size',       # S, M, L
    'company_location',   # country code
]
target = 'salary_in_usd'

print("Features we'll train on:")
for f in selected_features:
    unique_count = df[f].nunique()
    print(f"  {f}: {unique_count} unique values")

print(f"\nTarget: {target}")
print(f"Rows available: {len(df)}")
```
